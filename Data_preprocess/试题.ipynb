```python
import pandas as pd
df = pd.read_excel('data.xlsx', sheet_name = 'test', header=0, index_col='编号')
df.fillna(df.mean(), inplace=True)
data = df[(df.Zn > 80) & (df.Cu > 20)]
data = data.sort_values(by= 'Pb',ascending=True)
data.to_csv('result.csv')
data.iloc[:,0:4].plot(kind= 'box')
```

```python
import pandas as pd
df = pd.read_excel('data.xlsx', sheet_name = 'test', header=0, index_col='编号')
df.fillna(df.median(), inplace = True)
data = df[(df.Ni > 20) & (df.Co > 10)]
data = data.sort_values(by= 'Zn',ascending = False)
data.to_csv('result.csv')
data.iloc[:,0:6].plot(kind= 'bar')
```
