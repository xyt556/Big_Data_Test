```python
# 画出男女所占的比例，用饼图
# 男女的人数
m = 51212
f = 40742

# 算出男女所占的比例
m_perc = m/(m+f)
f_perc = f/(m+f)

# 指定颜色和标签
colors = ['navy','lightcoral']
labels = ['Male','Female']


# 指定画布大小
plt.figure(figsize = (6,6))
# 画出饼图 ：plt.pie()：先传入数据，指定标签，指定数据以百分比的形式输出，explode属性指定分离程度
paches,texts,autotexts = plt.pie([m_perc,f_perc],labels = labels,autopct = '%1.1f%%',explode = [0,0.05],
                                 colors = colors)

# 设置字体大小:text.set_fontsize
for text in texts+autotexts:
    text.set_fontsize(20)
    
    # 设置饼图里面的颜色： text.set_color
for text in autotexts:
    text.set_color('white')
```

```python
import numpy as np


x = np.linspace(0,10,1000)
y2 = np.sin(x**2)
y1 = x**2

# 指定外面的画布的大小，及坐标轴
fig,ax1 = plt.subplots()

# 指定子图的位置及大小
left,bottom,width,height = [0.2,0.35,0.3,0.23]

# 在外面画布的基础上指定子图的坐标轴，fig.add_axes(),添加轴坐标系
ax2 = fig.add_axes([left,bottom,width,height])

# 分别画出两个子图
ax1.plot(x,y1)
ax2.plot(x,y2)

plt.title('zhang')
```

```python
import matplotlib.pyplot as plt

# 定义该函数，在条形图上方添加值,模板
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        # 标注值的大小
        ax1.text(rect.get_x() + rect.get_width()/2.,1.02*height,'{:,}'.format(float(height)),ha = 'center',va = 'bottom',fontsize = 18)


top10_arrivals_countries = ['CANADA','NEXICO','UNITED\nKINGDOM','JAPAN','CHINA','GERMARY','SOUTH\nKOREA','FRANCE','BRAZIL','AUSTRELIA']
top10_arrivals_values = [16.625678,15.378026,3.934508,2.999718,2.618737,1.769498,1.628563,1.491409,1.2345345,1.12344345]
arrivals_countries = ['WESTERN\nEUROPE','ASIA','SOUTH\nAMERICA','OCEANIA','CARIBBEAN','MIDDLE\nEAST','CENTRAL\nAMERICA','EASTERN\nEUROPE','AFRICA']

# 转换成list 结构
labels = list(arrivals_countries)
print(labels)

# 百分比
arrivals_percent = [36.9,30.4,13.8,4.4,4.0,3.6,2.9,2.6,1.5]

fig,ax1 = plt.subplots(figsize = (20,12))

# 先画大图
zjy = ax1.bar(range(10),top10_arrivals_values,color = 'blue',alpha = 0.3)

# 对哪个坐标轴注释，就写在哪个坐标轴的后面，写x 轴的注释,子图没有xticks() 这个属性
plt.xticks(range(10),top10_arrivals_countries,fontsize = 18)



# 画子图，饼图
ax2 = inset_axes(ax1,width = 6,height = 6,loc = 5)
# 显示各个小饼的间隔属性explode
explode = (0.08,0.08,0.05,0.05,0.05,0.05,0.05,0.05,0.05)

# 画子图，百分比，标签，百分号显示，间隔
# ax2.pie(arrivals_percent,labels = arrivals_countries,autopct = '%1.1f%%',explode =explode)
patches,texts,autotexts = ax2.pie(arrivals_percent,autopct = '%1.1f%%',labels = labels,explode =explode,shadow=True, startangle=100)

# 设置子图的字体的大小
for text in texts+autotexts:
    text.set_fontsize(16)

# 将坐标轴 设置为不可见
for spine in ax1.spines.values():
    spine.set_visible(False)
                 
autolabel(zjy)
    
```

看一下就行，可以自己定义图形

```python
import numpy as np
from matplotlib.patches import Circle,Wedge,Polygon,Ellipse
from matplotlib.collections import PatchCollection
import matplotlib.pyplot as plt
%matplotlib inline

fig,ax = plt.subplots()

patches = []

leftstrip = Wedge((46,.5),.15,90,100)
midstrip = Wedge((.5,.5),.15,85,95)
rightstrip = Wedge((.54,.5),.15,80,90)
lefteye = Wedge((.36,.46),.06,0,360,width = 0.03)
righteye = Wedge((.63,.46),.06,0,360,width = 0.03)
nose = Wedge((.5,.32),.08,240,320,width = 0.03)
mouthleft = Wedge((.44,.4),.08,240,320,width = 0.01)
mouthright = Wedge((.56,.4),.08,220,300,width = 0.01)
patches += [leftstrip,midstrip,rightstrip,lefteye,righteye,nose,mouthleft,mouthright]

leftiris = Circle((.36,46),0.04)
rightiris = Circle((.63,46),0.04)
patches += [leftiris,rightiris]

leftear = Polygon([[.2,.6],[.4,.64]],True)
rightear = Polygon([[.6,.64],[.8,.6]],True)
topleftwhisker = Polygon([[.01,.4],[.18,.38],[17,.42]],True)
bottomleftwhisker = Polygon([[.01,.3],[.18,.32],[.2,.28]],True)
toprightwhisker = Polygon([[.99,.41],[.82,.39],[.82,.43]],True)
bottomrightwhisker = Polygon([[.99,.31],[.82,.33],[.81,.29]],True)
patches += [leftear,rightear,topleftwhisker,bottomleftwhisker,toprightwhisker,bottomrightwhisker]


body = Ellipse((0.5,-0.18),0.6,0.8)
patches.append(body)

colors = 100*np.random.rand(len(patches))
p = PatchCollection(patches,alpha = 0.4)
p.set_array(np.array(colors))
ax.add_collection(p)
```
