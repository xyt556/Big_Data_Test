# 分类属性绘图

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

%matplotlib inline

sns.set(color_codes = True)
np.random.seed(sum(map(ord,'regression')))

# tips = sns.load_dataset('./tips.csv')
tips = pd.read_csv('./tips.csv')
tips.head()

```

```python
# 结合树形图和小提琴图

sns.violinplot(x = 'day',y = 'total_bill',data = tips,inner = None)
sns.swarmplot(x = 'day',y = 'total_bill',data = tips,color = 'w',alpha = .5)
```

```python
sns.violinplot(x = 'day',y = 'total_bill',hue = 'sex',data = tips)
```

```python
sns.violinplot(x = 'day',y = 'total_bill',hue = 'sex',data = tips,split = True)
```

条形图： 显示值的集中趋势可以用条形图

```python
# 查看对于sex 的指标
sns.barplot(x = 'day',y = 'total_bill',data = tips,hue = 'sex')
```

点图可以更好地描述变化差异

```python
# 统计日期的变化情况，
sns.pointplot(x = 'day',y = 'total_bill',hue = 'sex',data = tips)
```

```python
sns.pointplot(x = 'day',y = 'total_bill',hue = 'sex',data = tips,
        palette = {"Male":"g","Female":"r"},markers =["^","o"],linestyles = ['-',"--"]);
```

```python
# 盒图横着画: 指定orient = "h"
# iris = sns.load_dataset("iris")
iris = pd.read_csv('./iris.csv')
sns.boxplot(data = iris,orient = "h")
```

```python
# 默认竖着画
sns.boxplot(data = iris)
```

多层面板分类图

```python
# 用facetorplot() 默认类型
sns.factorplot(x = 'day',y = 'total_bill',hue = "smoker",data = tips)
```

```python
# 指定kind 为bar
sns.factorplot(x = "day",y = "total_bill",hue = "smoker",data = tips,kind = "bar")
```

```python
# 指定个维度： col = "time"

sns.factorplot(x = "day",y = "total_bill",hue ="smoker",
               col = "time",data =tips,kind = "swarm")
```

```python
sns.factorplot(x = "time",y = "total_bill",hue = "smoker",col = "day",data = tips,
               kind = "box",height = 4,aspect = 0.5)
```

笔记：
- 第一点：
   kind:可选参数，默认point,
   bar 柱形图；
   count 频次；
   box 盒图
   violin 小提琴图；
   strip 散点；
   swarm 分散点；
   size 每个面的高度;
   aspect 纵横比；
   palette 调色板；
   seaborn 颜色色板或字典
   orient 方向“v/h”
