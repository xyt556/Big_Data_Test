#  子图与标注

```python
import numpy as np
import  pandas as pd
import matplotlib.pyplot as plt

zjy_np = np.arange(0,10,0.5)      # arange() 函数为numpy模块中的函数

# 画图plt.plot(x,y,字符颜色) 函数

plt.plot(zjy_np,zjy_np,'go')
plt.plot(zjy_np,zjy_np**2,'r-.')
plt.plot(zjy_np,zjy_np**3,'mp')

# 设置x,y 轴的标签

plt.xlabel('X',fontsize = 25)
plt.ylabel('Y',fontsize = 25)
```

```python
# 可以执行一条plt.plot() 函数，画多条数据

plt.plot(zjy_np,zjy_np,'go',zjy_np,zjy_np**2,'r-.',zjy_np,zjy_np**3,'mp')
```

指定线条宽度

```python
# 指定线条空间linspace（） 函数

x = np.linspace(-10,10)
y1 = np.cos(x)
y2 = np.sin(x)

plt.plot(x,y1,linewidth = 3.0)      # 指定线条宽度linewidth

# linestyle: 参数指定线条格式，
# marker : 参数指定一些标记点
#markerfacecolor : 参数指定标记点的颜色
# markersize : 指定标记点的大小

plt.plot(x,y2,color = 'b',linestyle = '-.',marker = '.',markerfacecolor = 'r',markersize = 15)
```

```python
# 可以先将图画出来，在设置参数，和格式，专门设置线的参数的函数为setp() 函数
# linewidth : 设置线宽度
#设置线透明程度 ：alpha ,为1的话，就是实线
line = plt.plot(x,y2)
plt.setp(line,color = 'g',linewidth = 2.0,alpha = 0.3)
```

子图

```python
# 子图横着排列

# 211 代表画的图是两行一列的，所站的位置是第一个

plt.subplot(211)
plt.plot(x,y2,color = 'g')

# 212 代表画的图是两行一列的所占的位置是第二个

plt.subplot(212)
plt.plot(x,y1,color = 'b')
```

```python
# 子图竖着排列

plt.subplot(121)
plt.plot(x,y2,color = 'r')

plt.subplot(122)
plt.plot(x,y1,color = 'b')
```

```python
# 画出的图为3 行2 列的图

plt.subplot(321)
plt.plot(x,y2,color = 'r')

plt.subplot(324)
plt.plot(x,y1,color = 'b')

plt.subplot(325)
plt.plot(x+1,y1,color = 'g')
```

```python
import numpy as np
import matplotlib 
matplotlib.use('nbagg') #如果要用到matplotlib.use() 必须在import matplotlib之后立即设置 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(0)
x = np.arange(5)
y = np.random.rand(5)

print(y)
plt.style.use('seaborn-darkgrid')
fig = plt.figure()                    # 定义fig，否者，错误：fig 未定义
fig,axes = plt.subplots(ncols = 2)     # fig.axes 错误：can't set attribute
v_bars = axes[0].bar(x,y,color = 'r')     # 以横轴为坐标轴
h_bars = axes[1].barh(x,y,color = 'r')    # 以竖轴为坐标轴
```

```python
np.random.seed(0)
x = np.arange(5)
y = np.random.randint(-5,5,5)

print(y)
plt.style.use('_classic_test')
# fig = plt.figure()                    # 定义fig，否者，错误：fig 未定义
fig,axes = plt.subplots(ncols = 2)     # fig.axes 错误：can't set attribute
v_bars = axes[0].bar(x,y,color = 'r')     # 以横轴为坐标轴
h_bars = axes[1].barh(x,y,color = 'r')    # 以竖轴为坐标轴

# axes[0].axhline() 函数：竖轴；    axes[1].axvline() 函数：横轴
# 将坐标轴画出来 ，加一条线,设置颜色为绿色，设置线宽度为2
axes[0].axhline(0,color = 'g',linewidth = 2)
axes[1].axvline(0,color = 'g',linewidth = 2)
```

```python
# 指定小于0 的值为绿色，大于0 的为红色

fig,axes = plt.subplots()
v_bars = axes.bar(x,y,color = 'lightblue')

for bar,height in zip(v_bars,y):
    if height < 0 :
        bar.set(edgecolor = 'darkred',color = 'g',linewidth = 3)

axes.axhline(0,color = 'r',linewidth = 3)       # 为坐标轴设置线
```
