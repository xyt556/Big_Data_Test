```python
cnData = pd.read_csv("../data/chinaData.csv")
```

```python
cnData
```

```python
m = folium.Map(location=[39, 106.5],
    tiles='Mapbox Bright',
    zoom_start=3
)

m.choropleth(
    geo_data="../data/CNGeojson.json",
    name='2009',
    data=cnData,
    columns=['FID', 'Pop_2009'],
    key_on='feature.id',
    fill_color='OrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='2009'
)

folium.LayerControl().add_to(m)

m
```

```python
import branca
```

```python
colorscale = branca.colormap.linear.PuBu_09.scale(0,10000)
```

```python
colorscale
```

```python
popSer = cnData.set_index('FID')['Pop_2009']

def style_function(feature):
    pop = popSer.get(int(feature['id']), None)
    return {
        'fillOpacity': 0.5,
        'weight': 0,
        'fillColor': '#black' if pop is None else colorscale(pop)
    }
```

```python
m = folium.Map(
    location=[39, 106],
    tiles='cartodb positron',
    zoom_start=3
)

folium.GeoJson(
    "../data/CNGeojson.json",
    name='geojson',
    style_function=style_function
).add_to(m)

m
```
