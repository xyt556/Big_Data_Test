```python
import folium
```

```python
cw = []
cwAttr = []
flag = True
with open("../data/天气.csv",'r',encoding='utf-8') as h:
    for i in h:
        if flag:
            flag = False
        else:
            line = i[0:-1].split(",")
            cwAttr.append(line)
            cw.append([float(line[-1]), float(line[-2])])
```

```python
print(cw[0])
print(cwAttr[0])
```

# 热度图（密度分析）

### min_opacity（默认值为1） - 热量开始时的最小不透明度。
### max_zoom（默认值为18） - 点数达到最大强度时的缩放级别（如缩放强度比例），默认等于默认地图的maxZoom
### max_val（float，默认值为1） - 最大点强度
### radius （int，默认值为25） - 热图中每个“点”的半径
### blur（int，默认值15） - 模糊量
### gradient （字典，默认无） - 颜色渐变配置。 例如 {0.4：'蓝色'，0.65：'lime'，1：'红色'}
### overlay （布尔，默认为True） - 将图层添加为可选叠加层（True）或基础层（False）。
### control （布尔，默认为True） - 图层是否将包含在LayerControls中。
### show（bool，默认为True） - 图层是否在开放时显示（仅适用于叠加层）。

```python
from folium.plugins import HeatMap
m = folium.Map([39, 106], tiles='stamentoner', zoom_start=3)
HeatMap(cw,
       radius =10).add_to(m)
m
```

```python
m = folium.Map([39, 106], tiles='stamentoner', zoom_start=3)
HeatMap(cw,
       radius=30).add_to(m)
m
```

# 点聚合

```python
from folium.plugins import MarkerCluster
m = folium.Map(location=[39, 105], zoom_start=5)
marker_cluster = MarkerCluster().add_to(m)

for c in range(300):
    folium.Marker(
        location=cw[c],
        popup="名称：{0}</br>最高温度：{1}</br>最低温度：{2}\
                天气情况：{3}</br>风向：{4}</br>风力：{5}\
        ".format(cwAttr[c][1],cwAttr[c][3],cwAttr[c][4],
                 cwAttr[c][5],cwAttr[c][6],cwAttr[c][7]),
        icon=folium.Icon(color='red', icon='remove-sign'),
    ).add_to(marker_cluster)
```
