```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as pml
import seaborn as sns
%matplotlib inline

from scipy import stats,integrate
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
# 函数FacetGrid() 函数： 可以展示数据中的很多子集
# 展示画图区域
g = sns.FacetGrid(tips,col = "time")
# 通过map()s 函数，实际的画图形,拿着画板变量g ,去画图
g.map(plt.hist,"tip")                 # 传进来的是plt.hist,画出条形图，
```

笔记：
- 第一点：map() 函数：
  第一个参数：要画成的图形类型；x,y 轴的坐标轴，透明度

```python
g = sns.FacetGrid(tips,col = "sex",hue = "smoker")

# plt.scatter: 画出散点图
g.map(plt.scatter,"total_bill","tip",alpha = 0.5)
g.add_legend()
```

```python
g = sns.FacetGrid(tips,row = "smoker",col = 'time',margin_titles = True)

# fit_reg = True, 画出回归的线    x_jitter： 指定一定的偏离程度
g.map(sns.regplot,"size","total_bill",color = "g",fit_reg = True,x_jitter = .1)
```

```python
# 设定一定的布局项， size 每个面的高度; aspect 纵横比；
g = sns.FacetGrid(tips,col = "day",height = 4,aspect = 0.5)
g.map(sns.barplot,"sex","total_bill")
```

```python
from pandas import Categorical
order_days = tips.day.value_counts().index
print(order_days)

# 自己指定画图的顺序
ordered_days = Categorical(['Thur', 'Fri', 'Sat', 'Sun'])

# FacetGrid：指定用pandas 格式：DataFrame 格式
g = sns.FacetGrid(tips,row= "day",row_order = ordered_days,height = 1.7,aspect = 2)
g.map(sns.boxplot,"total_bill",order =False)
```

# FacetGrid() 函数绘制多变量

```python
pal = dict(Lunch = 'green',Dinner = "red")
g = sns.FacetGrid(tips,hue = "time",palette = pal,height = 4)

# s = 50： 圆圈的大小
g.map(plt.scatter,"total_bill","tip",s = 50,alpha = 0.4,linewidth = 0.5,
      edgecolor = "white")
g.add_legend()
```

```python
g = sns.FacetGrid(tips,hue = "sex",palette = "Set1",height = 4,
                  hue_kws = {"marker":["^","o"]})

# s = 50： 圆圈的大小
g.map(plt.scatter,"total_bill","tip",s = 50,alpha = 0.4,linewidth = 0.5,
      edgecolor = "white")
g.add_legend()
```

```python
with sns.axes_style("white"):
    g = sns.FacetGrid(tips,row = "sex",col = "smoker",margin_titles = True,height = 4)
g.map(plt.scatter,"total_bill","tip",color= "g",edgecolor = 'white',lw = 0.5)
g.set_axis_labels("Total_bill","tips")
g.set(xticks = [10,30,50],yticks = [2,6,10])
g.fig.subplots_adjust(wspace = 0.62,hspace = 0.32)
```

```python
iris = sns.load_dataset("iris")
# 指定所要画的图的风格PairGrid
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
# 指定对角线上画什么图非对角线上画什么图
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
```

```python
# 指定对角线上画什么图非对角线上画什么图
# hue = 'species': 添加类别
g = sns.PairGrid(iris,hue = 'species')
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
```

```python
#画出特定特征的图：vars 指定
iris = sns.load_dataset("iris")
g = sns.PairGrid(iris,vars = ["sepal_length","sepal_width"],hue = "species")
g.map(plt.scatter)
```

```python
# 可以指定调色板的颜色
g = sns.PairGrid(tips,hue = "size",palette = "GnBu_d")
g.map(plt.scatter,s = 50,edgecolor = "white")
g.add_legend()
```

```python
np.random.seed(0)
normal_data = np.random.rand(3,3)
print(normal_data)

# center = 0: 以0 为中心画数据
ax = sns.heatmap(normal_data,center = 0)
```
