# 标签与数据格式的转换

```python
import numpy as np

# 将目标值y,找出来，转换为矩阵的形式
labels = np.array(df['actual'])

# 将数据集中的标签值列drop,生成x,也转换成矩阵的形式
features = df.drop('actual',axis = 1)

# 将名字单独保存一下,此步骤必须放在转换array之前
features_list = list(features.columns)

# AttributeError: 'numpy.ndarray' object has no attribute 'columns'

features = np.array(features)


print(features_list)
```

# 训练数据集与测试集

```python
# 指定随机的种子
seed = 42
np.random.seed(seed)

# 数据集的切分，test_size：代表的是切分的比例
from sklearn.model_selection import train_test_split

train_features,test_features,train_labels,test_labels = \
train_test_split(features,labels,test_size = 0.25,random_state = seed)
```

```python
print('训练集的特征：',train_features.shape)
print('测试集的特征：',test_features.shape)
print('训练集的标签：',train_labels.shape)
print('测试集的标签：',test_labels.shape)
```

# 建立一个基础的随机森林模型
- 先导入工具包，先建立1000个树试试，其他参数用默认值，之后再进行调参

```python
# 导入算法，指定的是随机森林的回归算法
from sklearn.ensemble import RandomForestRegressor

# 建立模型一定指定随机森林的树的个数：n_estimators
rf = RandomForestRegressor(n_estimators = 1000,random_state = seed)

# 训练模型
rf.fit(train_features,train_labels)
```

# 测试
-  先用MAPE指标来进行评估，也就是平均百分误差，回归任务有很多的评估方法，

```python
# 测试结果
predictions = rf.predict(test_features)

# 计算误差:
errors = abs(predictions - test_labels)

# mean absolute percentage error (MAPE)
# 为了避免除0的错误发生，加上一个小的误差浮动：0.01
mape = (errors / (test_labels + 0.01))

print("MAPE:",np.mean(mape))
```

# MAPE指标

```python
from sklearn.tree import export_graphviz
import pydot
```

```python
# 拿到一棵树
tree = rf.estimators_[5]

# 导出成dot文件
export_graphviz(tree,out_file = 'tree.dot',
                feature_names = features_list,
                rounded = True,
                precision = 1)

# 绘图
(graph,) = pydot.graph_from_dot_file('tree.dot')

# 展示
graph.write_png("tree.png");
```

```python
# 产看一下刚才的树的深度
print("树的深度是：",tree.tree_.max_depth)
```

```python
# 限制一下数的深度
rf_small = RandomForestRegressor(n_estimators = 10,max_depth = 3,random_state = seed)
rf_small.fit(train_features,train_labels)

# 提取一棵树
tree_small = rf_small.estimators_[5]

# 保存
export_graphviz(tree_small,out_file = 'small_tree.dot',
                   feature_names = features_list,
                  rounded = True,precision = 1)
(graph,) = pydot.graph_from_dot_file('small_tree.dot')

graph.write_png('small_tree.png');
```

# 特征重要性

```python
# 得到特征的重要性
importances = list(rf.feature_importances_)

# 转换格式
feature_importances = [(feature,round(importance,2)) for feature,importance in 
                       zip(features_list,importances)]

#排序
feature_importances = sorted(feature_importances,key = lambda x:x[1],reverse = True)

# 对应进行打印
[print('Variable:{:20} Imaportance:{}'.format(*pair)) for pair in feature_importances]
```

# 只用最重要的两个特征进行建立模型

```python
# 选择最重要的那两个特征
rf_most_important = RandomForestRegressor(n_estimators = 1000,random_state = seed)

#拿到着这两个特征
important_indices = [features_list.index('average'),features_list.index('temp_2')]
train_important = train_features[:,important_indices]
test_important = test_features[:,important_indices]

# 重新训练模型
rf_most_important.fit(train_important,train_labels)

# 预测结果
predictions = rf_most_important.predict(test_important)

errors = abs(predictions - test_labels)

# 评估结果

mape = np.mean(errors/(test_labels+0.01))
print("mape",mape)
```
