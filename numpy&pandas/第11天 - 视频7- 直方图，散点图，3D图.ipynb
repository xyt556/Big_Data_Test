```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 指定数据，高斯分布:均值0，标准差20，1000个数据
data = np.random.normal(0,20,1000)
bins = np.arange(-100,100,5)

# 画直方图 : 函数plt.hist(数据，bins)
plt.hist(data,bins = bins,color = 'g',alpha = 0.3)

# 指定x 轴的限制
plt.xlim(min(data)-5,max(data)+5)

# 添加网格
plt.grid()
```

笔记：
- 第一点：plt.legend()：函数添加标识
- 第二点：画直方图函数：plt.hist() 
       画条形图：bar(),barh(),
       画折线图plot() 
       小提琴图violinbar()
       盒图 boxplot()
       散点图 plt.scatter() 
- 第三点：plt.xlim() :指定x 轴的限制
- 第四点：plt.grid(): 指定网格

```python
import random

data1 = [random.gauss(15,10)for i in range(200)]
data2 = [random.gauss(5,5)for i in range(200)]

bins = np.arange(-50,50,2.5)

# 画直方图函数：hist() 画柱状图：bar(),barh(),画折线图plot() 小提琴图violinbar()

# 指定图的透明程度，可以看出多个图的透明程度
plt.hist(data1,bins = bins,label = 'class 1',alpha = 0.3)
plt.hist(data2,bins = bins,label = 'class 2',alpha = 0.3)

# 指定x 轴的限制
plt.xlim(min(min(data1),min(data2))-5,max(max(data1),max(data2))+5)

# plt.legend()：函数添加标识
plt.legend(loc = 'best')

# 将标识设置在上面
ax = plt.subplot(111)
ax.legend(loc = 'upper center',bbox_to_anchor = (0.9,1.2),ncol = 1)

# 添加网格
plt.grid()
```

# 散点图

```python
# 指定数据： 均值,协方差
mu_vecl = np.array([0,0])
cov_mat1 = np.array([[2,0],[0,2]])

x1_samples = np.random.multivariate_normal(mu_vecl,cov_mat1,100)
x2_samples = np.random.multivariate_normal(mu_vecl+0.2,cov_mat1+0.2,100)
x3_samples = np.random.multivariate_normal(mu_vecl+0.2*2,cov_mat1+0.2*2,100)

plt.figure(figsize = (8,6))

# 取坐标轴的二维坐标,传进来第一维数据，第二维数据，指定长什么样子，颜色，透明程度，标签
plt.scatter(x1_samples[:,0],x1_samples[:,1],marker = '^',color = 'red',alpha = 0.6,
            label = 'x1')
plt.scatter(x2_samples[:,0],x2_samples[:,1],marker = 'o',color = 'blue',alpha = 0.6,
            label = 'x2')
plt.scatter(x3_samples[:,0],x3_samples[:,1],marker = 'x',color = 'green',alpha = 0.6,
            label = 'x3')

plt.legend(loc = 'best')
plt.grid()
```

```python
x_data =[0.1,0.2,0.3,0.3,0.4,0.5,0.6]
y_data =[0.3,0.1,0.7,0.2,0.9,1,0.8]

fig = plt.figure(figsize = (8,6))
plt.scatter(x_data,y_data,marker = 'o',s = 30,color = 'r')

for x,y in zip(x_data,y_data):
# 标注坐标，用注释函数： plt.annotate(('(%s,%s)'%(x,y)： 获取坐标)，
#    xytext = (0,-15)：注释坐标的位置，
#   textcoords= 'offset points'：可以使坐标显示出来
    # ha = 'center' 可以使注释坐标对齐点
    plt.annotate('(%s,%s)'%(x,y),xy = (x,y),xytext = (0,-15),textcoords = 'offset points',
                ha = 'center')
```

```python
# 指定数据： 均值,协方差
mu_vecl = np.array([0,0])
cov_mat1 = np.array([[1,0],[0,1]])

X = np.random.multivariate_normal(mu_vecl,cov_mat1,500)
fig = plt.figure(figsize = (8,6))

# 指定离x轴远的点直径较大
R =X**2
R_sum = R.sum(axis = 1)

#画出散点图，定义参数marker 长得样子；s: 即为size大小
plt.scatter(X[:,0],X[:,1],marker = 'o',color = 'g',s = 20*R_sum)
```

# 3D图绘制

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

```python
x = np.arange(-4,4,0.25)
y = np.arange(-4,4,0.25)

fig = plt.figure()
# 将二维的画布转换为三维的画布用Axes3D() 函数
ax = Axes3D(fig)

# 将二维的x,y 的面转换成网格，z轴是在网格上进行画的
X,Y = np.meshgrid(x,y)
z = np.sin(np.sqrt(X**2+Y**2))
ax.plot_surface(X,Y,z,rstride = 1,cstride = 1,cmap = 'rainbow')

# 出现错误： Argument Z must be 2-dimensional.参数Z必须是二维的


    # 画出投影zdim 属性； 指定偏离程度：属性offset=2;  cmap
ax.contour(X,Y,z,zdim = 'z',offset = -2,cmap = 'rainbow')

# 限制z 轴坐标的高度-2->2，使投影更明显
ax.set_zlim(-2,2)
```

```python

# 绘制三维曲面
fig = plt.figure()
 
axes3d = Axes3D(fig)
 
#!！面
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
 
X,Y = np.meshgrid(x,y)
Z = np.sqrt(X**2+Y**2)
 
axes3d.plot_surface(X,Y,Z)
```

# 3D中的散点图

```python
np.random.seed(1)

def randrange(n,vmin,vmax):
    return (vmax - vmin)*np.random.rand(n)+vmin

fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')

n = 100

for c,m,zlow,zhigh in [('r','o',-50,-30),('g','^',-30,-5)]:
    xs = randrange(n,23,32)
    ys = randrange(n,0,100)
    zs = randrange(n,zlow,zhigh)
    ax.scatter(xs,ys,zs,c=c,marker = m)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')

for c,z in zip(['r','g','b','y'],[30,20,10,0]):
    xs = np.arange(20)
    ys =np.random.rand(20)
    
    # 指定颜色
    cs = [c]*len(xs)
    # 可以指定zdir = ’y‘属性，让坐标轴为xOy面，即使条形图立起来
    ax.bar(xs,ys,zs = z,zdir = 'y',color = cs,alpha = 0.8)
    
```
