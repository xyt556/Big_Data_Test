# 盒图绘制

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['SimHei']
mpl.rcParams['axes.unicode_minus'] = True
```

```python
# 构造数据，均值：0；方差：std;构造100个数据
zjy_data = [np.random.normal(0,std,100)for std in range(1,4)]
# 设置画布的大小
fig = plt.figure(figsize = (8,6))
# 话盒图的函数plt.boxplot()函数，sym 属性设置：s ,用方框表示数据
plt.boxplot(zjy_data,notch = False,sym= 's',vert = True)

# 设置x轴的刻度函数plt.xticks()
plt.xticks([y+1 for y in range(len(zjy_data))],['X1','X2','X3'])
plt.xlabel('x',fontsize = 20)
plt.title('张济源的盒图')
```

# 盒图细节

```python
# 构造数据，均值：0；方差：std;构造100个数据
zjy_data = [np.random.normal(0,std,100)for std in range(1,4)]
# 设置画布的大小
fig = plt.figure(figsize = (8,6))
# 盒图的函数plt.boxplot()函数，sym 属性设置：s ,用方框表示数据；vert属性表示竖着画：True
bplot = plt.boxplot(zjy_data,notch = False,sym= 's',vert = True,patch_artist = True)


# 设置x轴的刻度函数plt.xticks()
plt.xticks([y+1 for y in range(len(zjy_data))],['X1','X2','X3'])
plt.xlabel('x',fontsize = 20)
plt.title('张济源的盒图')

# for components in bplot.keys():
#     for line in bplot[components]:
#         line.set_color('black')

# 进行颜色的填充
color = ['pink','lightblue','lightgreen']
for patchch,color in zip(bplot['boxes'],color):
    patchch.set_facecolor(color)
    
# 报错： AttributeError: 'Line2D' object has no attribute 'set_facecolor'
# 解决：plt.boxplot(patch_artist = True)
```

# 小提琴图violinplot

```python
zjy_data = [np.random.normal(0,std,100)for std in range(6,10)]
fig,axes = plt.subplots(nrows = 1,ncols = 2,figsize = (8,6))
# 画第一个图
axes[0].violinplot(zjy_data,showmeans = False,showmedians = True)
axes[0].set_title('violin plot')

axes[1].boxplot(zjy_data)
axes[1].set_title('box plot')

# 对每一个图都设置一下网格
for ax in axes:
    ax.yaxis.grid(True)
    ax.set_xticks([y+1 for y in range(len(zjy_data))])
    
# 设置x 轴的刻度及标签
plt.setp(axes,xticks = [y+1 for y in range(len(zjy_data))],
         xticklabels = ['x1','x2','x3','x4'])
```

# 绘图细节的设置

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

# 构造数据
x = range(10)
y = range(10)

# 得到画图的东西，画布
fig = plt.gca()

# 将（x,y）画出来
plt.plot(x,y)

# 去掉 x轴的刻度
fig.axes.get_xaxis().set_visible(False)

# 去掉Y轴的刻度
fig.axes.get_yaxis().set_visible(False)
```

```python
import math
x = np.random.normal(loc = 0.0,scale = 1.0,size = 300)
width = 0.5

# 设置间隔，floor()函数 向下取整，ceil() 函数向上取整
bins = np.arange(math.floor(x.min())-width,math.ceil(x.max())+width,width)
ax = plt.subplot(111)

# 此处的subplot() 函数不能改为subplots() 函数，会出现，元组没有spines() 这个属性
# 设置轴不可见
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# 设置成轴刻度不可见
# plt.tick_params(bottom = 'on',right = 'off',top = 'off')

# 添加网格
plt.grid()

#画直方图 plt.hist() 以x 为轴，alpha 透明度，bins 间隔
plt.hist(x,alpha = 0.5,bins = bins)
```

```python
x = range(10)
y = range(10)

labels = ['zhang' for i in range(10)] 
fig,ax = plt.subplots()
plt.plot(x,y)

# 设置x轴的刻度标签：函数set_xticklabels（），
# 设置对齐参数:horizontalalignment,设置旋转角度：rotation 
ax.set_xticklabels(labels,rotation = 45,horizontalalignment = 'left')
```

# 绘图细节设置-2

```python
# 创建数组：含有等距的10个元素
x = np.arange(10)
# 画3 条线
for i in range(1,4):
    plt.plot(x,i*x**2,label = 'Group %d'%i)

# 添加标识，可以设置参数loc = 'best',会自动找一下最好的位置，不会使标识挡着数据
plt.legend(loc = 'best')
```

```python
# 做数据
x = np.arange(10)

# 设置画布
fig = plt.figure()
ax = plt.subplot(111)

for i in range(1,4):
    plt.plot(x,i*x**2,label = 'Group %d'%i)
    
# 用ax.legend() 将标识设置在外面，参数：bbox_to_anchor：设置
# plt.legend() 将标识设置在里面
ax.legend(loc = 'upper center',bbox_to_anchor = (0.9,1.3),ncol = 1)
```
